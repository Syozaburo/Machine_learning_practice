```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
from IPython.display import display

import warnings
warnings.filterwarnings('ignore')

#サンプルデータの読み込み
df_iris = sns.load_dataset("iris")

#項目名の日本語化
columns_i = ['がく片長', 'がく片幅', '花弁長', '花弁幅', '種別']
df_iris.columns = columns_i

#データの内容
display(df_iris.head())
```

```python
#散布図表示

#グラフ描画領域の調整
plt.figure(figsize=(6, 6))

#散布図の表示
plt.scatter(df_iris['がく片長'], df_iris['花弁長'])

#ラベル表示
#sepal がく petal 花弁
plt.xlabel('がく片長')
plt.ylabel('花弁長')
plt.show()
```

```python
#散布図表示 seaborn版(1)

plt.figure(figsize=(6, 6))
sns.scatterplot(x='がく片幅', y='花弁長', hue='種別', s=70, 
                data=df_iris)
plt.show()
```

```python
#全散布図同時表示

sns.pairplot(df_iris, hue='種別')
plt.show()
```

```python
#散布図表示 seaborn版(2)

sns.jointplot(x='がく片幅', y='花弁長', data=df_iris)
plt.show()
```

```python
#箱ひげ図表示 matplotlib版

#グラフ描画領域の調整
plt.figure(figsize=(6, 6))

#箱ひげ図の描画
df_iris.boxplot(patch_artist=True)
plt.show()
```

```python
# melt関数によるデータ事前加工
w = pd.melt(df_iris, id_vars=['種別'])

#加工結果の確認
display(w.head())
```

```python
#seabornによる箱ひげ図の表示

#hueパラメータを追加し、花の種類で箱ひげ図を書き分ける
plt.figure(figsize=(6, 6))
sns.boxplot(x="variable", y="value", data=w, hue='種別')
plt.show()
```
